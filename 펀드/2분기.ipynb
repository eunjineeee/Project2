```python
kospi_1 = pd.read_csv('/content/drive/MyDrive/PJ2/펀드 위험지수/csv/2분기/코스피_반기_개별.csv', encoding = 'euc-kr')
kospi_2 = pd.read_csv('/content/drive/MyDrive/PJ2/펀드 위험지수/csv/2분기/코스피_반기_연결.csv', encoding = 'euc-kr')
```

```python
# 연결 개별 데이터 프레임 합치기 (회사명, 거래소코드, 회계년도를 기준으로)
KOSPI = kospi_2.merge(kospi_1, on = ['회사명', '거래소코드', '회계년도'], how = 'left')
```

```python

#  연결 재무제표를 개별재무제표로 대체하는 함수
def replace_yeongyeol_with_gaebyeol(df):
    connected_cols = [col for col in df.columns if "연결" in col]

    for connected_col in connected_cols:
        original_col = connected_col.replace("연결", "")
        if original_col in df.columns:
            df.loc[df[connected_col].isnull(), connected_col] = df[original_col]
    return df
```

```python
replace_yeongyeol_with_gaebyeol(KOSPI)
```

```python
KOSPI.columns
```

```python
# 결측값 214개 확인
replace_yeongyeol_with_gaebyeol(KOSPI).isna().sum()
```

```python
# 개별재무제표 컬럼 삭제
KOSPI.drop(columns = ['자산(*)(요약)(IFRS연결)(백만원)',
       '유동부채(*)(요약)(IFRS연결)(백만원)', '비유동부채 (*)(요약)(IFRS연결)(백만원)',
       '부채(*)(요약)(IFRS연결)(백만원)', '자본(*)(요약)(IFRS연결)(백만원)',
       '재고자산(요약)(IFRS연결)(백만원)', '당기순이익(총포괄손익)(요약)(IFRS연결)(백만원)',
       '매출액(수익)(요약)(IFRS연결)(백만원)', '판매비와 관리비(물류원가 등 포함)(요약)(IFRS연결)(백만원)',
       '매출원가(요약)(IFRS연결)(백만원)', '매입채무(IFRS연결)(천원)', '금융원가(요약)(IFRS연결)(백만원)'], axis = 1, inplace = True)
```

```python
# 개별 재무제표 컬럼 삭제 위해 컬럼명 추출
KOSPI.columns
```

```python
KOSPI.isna().sum()
```

```python
KOSPI['매입채무(IFRS)(천원)']/1000
```

```python
# () 앞에 있는 이름만 남기고 나머지 삭제하는 함수
def extract_name(KOSPI):
    return KOSPI.split('(')[0]

# 데이터프레임의 모든 컬럼에 대해 () 앞에 있는 이름만 남기고 나머지 삭제
KOSPI.columns = KOSPI.columns.map(extract_name)
```

```python
KOSPI.rename(columns={'금융원가': '금융비용'}, inplace=True)
```
