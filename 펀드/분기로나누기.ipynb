## 당기순이익

```python
# 년도 컬럼 추출 위해 데이터 타입 변경
df_1분기["회계년도"] = pd.to_datetime(df_1분기["회계년도"])
df_2분기["회계년도"] = pd.to_datetime(df_2분기["회계년도"])
df_2분기
```

```python
df_1분기
```

```python
# 회계년 컬럼 생성
df_2분기['회계년'] = df_2분기['회계년도'].dt.year
df_2분기
```

```python
df_1분기['회계년'] = df_1분기['회계년도'].dt.year
df_1분기
```

```python
# 2분기 데이터프레임에 1분기 값 merge
df_2분기 = df_2분기.merge(df_1분기[['거래소코드','회계년', '당기순이익']], on=['거래소코드', '회계년'], suffixes=('_2분기', '_1분기'))
df_2분기
```

```python
# 당기순이익 컬럼 생성
df_2분기['당기순이익'] = df_2분기['당기순이익_2분기'] - df_2분기['당기순이익_1분기']
```

```python
df_2분기
```

## 매출액

```python
# 2분기 데이터프레임에 1분기 값 merge
df_2분기 = df_2분기.merge(df_1분기[['거래소코드','회계년', '매출액']], on=['거래소코드', '회계년'], suffixes=('_2분기', '_1분기'))
df_2분기
```

```python
# 매출액 컬럼 생성
df_2분기['매출액'] = df_2분기['매출액_2분기'] - df_2분기['매출액_1분기']
```

## 판관비

```python
# 2분기 데이터프레임에 1분기 값 merge
df_2분기 = df_2분기.merge(df_1분기[['거래소코드','회계년', '판매비와 관리비']], on=['거래소코드', '회계년'], suffixes=('_2분기', '_1분기'))
df_2분기
```

```python
# 판관비 컬럼 생성
df_2분기['판매비와 관리비'] = df_2분기['판매비와 관리비_2분기'] - df_2분기['판매비와 관리비_1분기']
```

## 매출원가

```python
# 2분기 데이터프레임에 1분기 값 merge
df_2분기 = df_2분기.merge(df_1분기[['거래소코드','회계년', '매출원가']], on=['거래소코드', '회계년'], suffixes=('_2분기', '_1분기'))
df_2분기
```

```python
# 매출원가 컬럼 생성
df_2분기['매출원가'] = df_2분기['매출원가_2분기'] - df_2분기['매출원가_1분기']
```

## 금융비용

```python
# 2분기 데이터프레임에 1분기 값 merge
df_2분기 = df_2분기.merge(df_1분기[['거래소코드','회계년', '금융비용']], on=['거래소코드', '회계년'], suffixes=('_2분기', '_1분기'))
df_2분기
```

```python
# 매출원가 컬럼 생성
df_2분기['금융비용'] = df_2분기['금융비용_2분기'] - df_2분기['금융비용_1분기']
```

```python
df_2분기.columns
```

```python
# 필요없는 컬럼 삭제
df_2분기.drop(columns = ['당기순이익_2분기', '매출액_2분기', '판매비와 관리비_2분기', '매출원가_2분기','금융비용_2분기', '회계년', '당기순이익_1분기',
             '매출액_1분기', '판매비와 관리비_1분기', '매출원가_1분기', '금융비용_1분기'], inplace = True)
```
