## 1분기 비율 컬럼 생성

```python
df_1분기.info()
```

- 피쳐 : 1. 순운전자본비율 ~ 10. 총자산투자효율

```python
# 1. 순운전자본비율 피쳐 생성
df_1분기['순운전자본비율'] = ((df_1분기['유동자산'] - df_1분기['유동부채'])/df_1분기['자산']) * 100
```

```python
## 유동부채가 음수라면?
df_1분기[df_1분기['유동부채'] < 0]
```

```python
## 유동자산 - 유동부채랑 자산 둘다 음수라면?
df_1분기[((df_1분기['유동자산'] - df_1분기['유동부채'])<0) & (df_1분기['자산'] < 0)]
```

```python
## 유동자산 - 유동부채랑 자산 둘다 음수라면?
df_1분기[((df_1분기['유동자산'] - df_1분기['유동부채'])<0) & (df_1분기['자산'] < 0)]
```

```python
# 2.부채비율
## 자기자본 컬럼 생성
df_1분기['자기자본'] = df_1분기['자산'] - df_1분기['부채']
```

```python
## 부채가 음수라면?
df_1분기[df_1분기['부채']<0]
```

```python
## 부채비율 컬럼 생성
df_1분기['부채비율'] = (df_1분기['부채'] / df_1분기['자기자본']) * 100
```

```python
## 둘 다 음수라면?
df_1분기[(df_1분기['자기자본']<0) & (df_1분기['부채'] < 0)]
```

```python
# 3. 재고자산증가율 컬럼 생성
df_1분기['재고자산증가율'] = df_1분기['재고자산'].pct_change() * 100
df_1분기['재고자산증가율'].iloc[0] = 0.0
```

```python
# 4. 총자산증가율 컬럼 생성
df_1분기['총자산증가율'] = df_1분기['자산'].pct_change() * 100
df_1분기['총자산증가율'].iloc[0] = 0.0
```

```python
# 5. 매입채무회전기간 컬럼 생성
df_1분기['매입채무회전기간'] = 365 / (df_1분기['매출원가'] / df_1분기['매입채무'])
```

```python
## 둘 다 음수라면?
df_1분기[(df_1분기['매출원가']<0) & (df_1분기['매입채무']<0)]
```

```python
# 6. 매출액순이익률 컬럼 생성
df_1분기['매출액순이익률'] = (df_1분기['당기순이익'] / df_1분기['매출액']) * 100
```

```python
## 둘 다 음수라면?
df_1분기[(df_1분기['매출액']<0) & (df_1분기['당기순이익']<0)]
```

```python
# 7. 자기자본순이익률 컬럼 생성
df_1분기['자기자본순이익률'] = (df_1분기['당기순이익'] / df_1분기['자기자본']) * 100
```

```python
## 둘 다 음수라면?
df_1분기[(df_1분기['자기자본']<0) & (df_1분기['당기순이익']<0)]
```

```python
## 음수로 변경
df_1분기.loc[(df_1분기['자기자본'] < 0) & (df_1분기['당기순이익'] < 0), '자기자본순이익률'] *= -1
```

```python
# 8. 금융비용 대 총비용비율 컬럼 생성
df_1분기['금융비용 대 총비용비율'] = (df_1분기['금융비용'] / (df_1분기['매출원가'] + df_1분기['판매비와 관리비'])) * 100
```

```python
## 둘 다 음수라면?
df_1분기[(df_1분기['금융비용']<0) & ((df_1분기['매출원가'] + df_1분기['판매비와 관리비'])<0)]
```

```python
# 9. 총자산사업이익률 컬럼 생성
df_1분기['총자산사업이익률'] = (df_1분기['당기순이익'] / df_1분기['자산']) * 100
```

```python
## 둘 다 음수라면?
df_1분기[(df_1분기['당기순이익']<0) & (df_1분기['자산']<0)]
```

```python
df_1분기['부가가치'] = df_1분기['매출액'] - (df_1분기['매출원가'] + df_1분기['판매비와 관리비'])
```

```python
# 10. 총자산투자효율
df_1분기['총자산투자효율'] = (df_1분기['부가가치'] / df_1분기['자산']) * 100
```

```python
## 둘 다 음수라면?
df_1분기[(df_1분기['부가가치']<0) & (df_1분기['자산']<0)]
```
