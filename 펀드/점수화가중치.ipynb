## 2022년 베타값

```python
beta = pd.read_csv('./코스피,코스닥 52주베타(2022.01~2022.12).csv')
beta
```

```python
beta.columns = ['거래소코드', '베타계수', 0, 1, 2,  3 , 4]
```

```python
beta.drop(columns = [0, 1, 2, 3, 4], inplace = True)
```

```python
beta.info()
```

```python
beta
```

```python
beta[beta['거래소코드'].str.contains('K')]
```

```python
# '거래소코드' 컬럼에서 'K' 값을 포함하는 행을 없애기
beta['거래소코드'] = beta['거래소코드'].str.replace('K', '')
beta['거래소코드'] = beta['거래소코드'].str.replace('L', '')
beta['거래소코드'] = beta['거래소코드'].str.replace('M', '')
beta['거래소코드'] = beta['거래소코드'].str.replace(':', '')
beta['거래소코드'] = beta['거래소코드'].str.replace(' ', '')
```

```python
# 숫자로 변환 
beta['거래소코드'] = beta['거래소코드'].astype(int)
```

```python
beta
```

```python
score['회계년도'] = pd.to_datetime(score['회계년도'])
score_log['회계년도'] = pd.to_datetime(score_log['회계년도'])
score_num['회계년도'] = pd.to_datetime(score_num['회계년도'])
```

```python
score['회계년'] = score['회계년도'].dt.year
score_log['회계년'] = score_log['회계년도'].dt.year
score_num['회계년'] = score_num['회계년도'].dt.year
```

## 베타계수 merge

```python
score.shape
```

```python
score = score.merge(beta, on='거래소코드')
```

```python
score.loc[score['회계년'] != 2022, '베타계수'] = None
```

```python
score_log = score_log.merge(beta, on='거래소코드')
```

```python
score_log.loc[score_log['회계년'] != 2022, '베타계수'] = None
```

```python
score_num = score_num.merge(beta, on='거래소코드')
```

```python
score_num.loc[score_num['회계년'] != 2022, '베타계수'] = None
```
