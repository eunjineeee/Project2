```python
import pandas as pd
import numpy as np
```

```python
score_log = pd.read_csv('./점수화_상용로그 편.csv')
score_log
```

```python
score_log.drop(columns = ['베타계수', '기업별 표준편차'], inplace = True)
```

```python
score_log['회계년도'] = pd.to_datetime(score_log['회계년도'])
```

```python
score_log.info()
```

```python
score_log['분기별 기업 위험 점수'].describe()
```

```python
score_log['회계달'] = score_log['회계년도'].dt.month
```

## 기간 2022년 12월 기준

```python
score_log_22_12 = score_log[(score_log['회계년'] == 2022) & (score_log['회계달'] == 12)]
```

```python
score_log_22_12 = score_log_22_12.sample(n=50, random_state=42)
```

```python
score_log_50개기업 = score_log_22_12.sort_values(by='분기별 기업 위험 점수', ascending=False)
score_log_50개기업
```

```python
score_log_50개기업.reset_index(drop = True, inplace = True)
```

```python
score_log_50개기업['거래소코드'].unique()
```

```python
거래소코드 = [373200,  27970,  31510, 224060, 235980,  41650, 139990,  32300,
        82740,  90350,  13360,  11760, 214680,  51500,  14910,  50110,
         1040,  84990,  16100, 368600, 340360, 272550,  15540,  89850,
        51380, 288620,  14200,  58730,  25440,   7340,  81000,  36010,
        39130,   7680,  33780, 220180,  60380,  18310,   1880, 216050,
       377300, 306200, 359090,  37070, 230240,  15230, 228340,  84010,
       205100, 111710]
```

```python
score_log_전체50 = score_log[score_log['거래소코드'].isin(거래소코드)]
```

```python
score_log_전체50 = score_log_전체50[(score_log_전체50['회계년'] == 2022) | (score_log_전체50['회계년'] == 2021)]
```

```python
score_log_전체50
```

```python
# '거래소코드'와 '회계년'이 같은 경우의 '분기별 기업 위험점수' 평균 계산
grouped_df = score_log_전체50.groupby(['거래소코드', '회계년'])['분기별 기업 위험 점수'].mean().reset_index()

# 평균 값을 '평균 위험점수' 컬럼으로 추가
score_log_전체50 = score_log_전체50.merge(grouped_df, on=['거래소코드', '회계년'], suffixes=('', '_평균'), how='left')
score_log_전체50
```

```python
score_log_전체50.drop_duplicates(subset=['거래소코드', '회계년'], keep='first', inplace=True)
```

```python
score_log_전체50
```

```python
# 가중치를 계산하여 새로운 컬럼으로 추가
score_log_전체50['분기별 기업 위험 점수_가중치'] = score_log_전체50.apply(lambda row: 0.7 if row['회계년'] == 2022 else 0.3, axis=1) * score_log_전체50['분기별 기업 위험 점수_평균']

# 컬럼 추가 후 데이터프레임 출력
score_log_전체50
```

```python
# 가입 시점 점수
print("가입 시점 펀드 위험 지수 :", round(score_log_전체50['분기별 기업 위험 점수_가중치'].sum(),3))
```

```python
# 현재 시점 점수
print("현재 시점 펀드 위험 지수 :",round(score_log_50개기업['분기별 기업 위험 점수'].sum(),3))
```
