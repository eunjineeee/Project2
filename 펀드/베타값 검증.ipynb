```python
import FinanceDataReader as fdr # 52주 베타값

# Function to calculate beta
def calculate_beta(stock_returns, market_returns):
    covariance = stock_returns.cov(market_returns)
    market_variance = market_returns.var()
    beta = covariance / market_variance
    return beta

# 코스피 상장 기업들의 거래소 코드 리스트
kospi_tickers = [219750, 156100,  30790, 203690,  82210,  58450,    230, 249420,
         1210, 304840, 217330, 217480, 301300,  83660,  60900,  72520,
       177350,  80530,  49120, 122350,   8500, 196300,  65170,  69140,
         6380, 100120,  11200, 273640,  78000, 352910,  99390, 147830,
       290670, 403870, 275630, 366030, 219550, 336060, 241770, 340570,
       347740,  94970,  94360, 179290, 335890,  14570,  64760, 376980,
        58630, 153460]


# 6자리로 만들기 위해 코드 앞에 0을 붙여줍니다.
kospi_tickers = [str(code).zfill(6) for code in kospi_tickers]

# 2021년 1월 1일부터 2021년 12월 31일까지의 주가 데이터를 가져올 날짜 범위
start_date = '2022-06-01'
end_date = '2023-5-20'

# Load the KOSPI stock listing data only once outside the loop
kospi_listing = fdr.StockListing('KOSPI')

# Store beta values in a dictionary
all_betas = {}

# 50개 기업들의 2021년 52주 베타값 계산
for ticker in kospi_tickers:
    try:
        # 기업의 주가 데이터 가져오기
        stock_data = fdr.DataReader(ticker, start_date, end_date)

        # 일간 수익률 계산
        stock_returns = stock_data['Close'].pct_change().dropna()
        market_returns = fdr.DataReader('KS11', start_date, end_date)['Close'].pct_change().dropna()

        # 2021년 52주 베타값 계산
        beta_2021 = calculate_beta(stock_returns, market_returns)
        all_betas[ticker] = beta_2021
    except Exception as e:
        print(f"Error processing {ticker}: {e}")

# 결과 출력
print("----- 코스피 상장 기업의 2021년 52주 베타값 -----")
for ticker, beta in all_betas.items():
    print(f"{ticker}: {beta:.4f}")

# 결과를 DataFrame으로 변환
result_df = pd.DataFrame.from_dict(all_betas, orient='index', columns=['Beta'])
```

# 표준편차

```python
import FinanceDataReader as fdr # 표준편차 구하는 코드
import pandas as pd

# Function to calculate standard deviation
def calculate_std_dev(stock_returns):
    std_dev = stock_returns.std()
    return std_dev

# 코스피 상장 기업들의 거래소 코드 리스트
kospi_tickers = [219750, 156100,  30790, 203690,  82210,  58450,    230, 249420,
         1210, 304840, 217330, 217480, 301300,  83660,  60900,  72520,
       177350,  80530,  49120, 122350,   8500, 196300,  65170,  69140,
         6380, 100120,  11200, 273640,  78000, 352910,  99390, 147830,
       290670, 403870, 275630, 366030, 219550, 336060, 241770, 340570,
       347740,  94970,  94360, 179290, 335890,  14570,  64760, 376980,
        58630, 153460]


# 6자리로 만들기 위해 코드 앞에 0을 붙여줍니다.
kospi_tickers = [str(code).zfill(6) for code in kospi_tickers]

# 2022년 1월 1일부터 2022년 12월 31일까지의 주가 데이터를 가져올 날짜 범위
start_date = '2022-06-01'
end_date = '2023-05-20'

# Store standard deviation values in a dictionary
all_std_devs = {}

# 50개 기업들의 2022년 표준편차 계산
for ticker in kospi_tickers:
    try:
        # 기업의 주가 데이터 가져오기
        stock_data = fdr.DataReader(ticker, start_date, end_date)

        # 일간 수익률 계산
        stock_returns = stock_data['Close'].pct_change().dropna()

        # 2022년 표준편차 값 계산
        std_dev_2022 = calculate_std_dev(stock_returns)
        all_std_devs[ticker] = std_dev_2022
    except Exception as e:
        print(f"Error processing {ticker}: {e}")

# 결과 출력
print("----- 코스피 상장 기업들의 2022년 표준편차 -----")
for ticker, std_deviation in all_std_devs.items():
    print(f"{ticker}: {std_deviation:.4f}")
# 결과를 DataFrame으로 변환
result_df = pd.DataFrame.from_dict(all_std_devs, orient='index', columns=['Std_Deviation'])
```

# 개별 기업 52주 베타값 정상 출력되는지 검증

```python
import FinanceDataReader as fdr

# 삼성전자의 2022년 1월 1일부터 2022년 12월 31일까지의 주가 데이터 가져오기
start_date = '2021-01-01'
end_date = '2021-12-31'

samsung_data = fdr.DataReader('257720', start_date, end_date)

# 52주 베타값 계산 함수
def calculate_beta(stock_data, market_data):
    # 일간 수익률 계산
    stock_returns = stock_data['Close'].pct_change().dropna()
    market_returns = market_data['Close'].pct_change().dropna()

    # 베타값 계산
    covariance = stock_returns.cov(market_returns)
    market_variance = market_returns.var()
    beta = covariance / market_variance

    return beta

# 2022년 삼성전자의 52주 베타값 계산을 위해 코스피 지수 데이터 가져오기
kospi_data = fdr.DataReader('KS11', start_date, end_date)

# 2022년 삼성전자의 52주 베타값 계산
beta_2022 = calculate_beta(samsung_data, kospi_data)

# 결과 출력
print("삼성전자 2022년 52주 베타값:", beta_2022)
```

```python
import FinanceDataReader as fdr # 전체기업 베타값

# 2021년 1월 1일부터 2021년 12월 31일까지의 주가 데이터를 가져올 날짜 범위
start_date = '2021-01-01'
end_date = '2021-12-31'

# 52주 베타값 계산 함수
def calculate_beta(stock_returns, market_returns):
    covariance = stock_returns.cov(market_returns)
    market_variance = market_returns.var()
    beta = covariance / market_variance
    return beta

# 50개의 기업들의 2021년 52주 베타값을 저장할 딕셔너리
all_betas = {}

# 코스피 지수(KOSPI) 데이터 가져오기
kospi_data = fdr.DataReader('KS11', start_date, end_date)

# KOSPI 상장 기업들의 코드 리스트
kospi_tickers = fdr.StockListing('KOSPI')['Code'].tolist()

# 50개 기업들의 2021년 52주 베타값 계산
for ticker in kospi_tickers:
    try:
        # 기업의 주가 데이터 가져오기
        stock_data = fdr.DataReader(ticker, start_date, end_date)

        # 일간 수익률 계산
        stock_returns = stock_data['Close'].pct_change().dropna()
        market_returns = kospi_data['Close'].pct_change().dropna()

        # 2021년 52주 베타값 계산
        beta_2021 = calculate_beta(stock_returns, market_returns)
        all_betas[ticker] = beta_2021
    except Exception as e:
        print(f"Error processing {ticker}: {e}")

# 결과 출력
print("----- 코스피 상장 기업의 2021년 52주 베타값 -----")
for ticker, beta in all_betas.items():
    # Retrieve the company name from the loaded StockListing data
    company_name = fdr.StockListing('KOSPI').loc[fdr.StockListing('KOSPI')['Code'] == ticker, 'Name'].values
    if len(company_name) > 0:
        print(f"{company_name[0]} ({ticker}): {beta:.4f}")
    else:
        print(f"Unknown Company ({ticker}): {beta:.4f}")
```
