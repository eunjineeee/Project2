```python
KOSPI_1분기_연결 = pd.read_csv('./datasets2/코스피_1분기_연결.csv', encoding = 'euc-kr')
KOSPI_1분기_개별 = pd.read_csv('./datasets2/코스피_1분기_개별.csv', encoding = 'euc-kr')
KOSPI_1분기_연결
```

```python
# 연결 개별 데이터 프레임 합치기 (회사명, 거래소코드, 회계년도, 상장일, 상장폐지일, 소속코드를 기준으로)
KOSPI = KOSPI_1분기_연결.merge(KOSPI_1분기_개별, on = ['회사명', '거래소코드', '회계년도'], how = 'left')
```

```python
KOSPI.info()
```

```python
KOSPI.isna().sum()
```

```python
#  연결 재무제표를 개별제무제표로 대체하는 함수
def replace_yeongyeol_with_gaebyeol(df):
    connected_cols = [col for col in df.columns if "연결" in col]

    for connected_col in connected_cols:
        original_col = connected_col.replace("연결", "")
        if original_col in df.columns:
            df.loc[df[connected_col].isnull(), connected_col] = df[original_col]

    return df
```

```python
# 회사명과 거래소코드 데이터프레임 생성

회사명KOSDAQ = KOSDAQ_1분기_연결[["회사명","거래소코드"]]
회사명KOSDAQ = 회사명KOSDAQ.groupby("회사명")["거래소코드"].unique()
회사명KOSDAQ = pd.DataFrame(회사명KOSDAQ)
회사명KOSDAQ
```

```python
# 동일 회사명 이름 확인
for i in 회사명KOSDAQ.index:
    if len(회사명KOSDAQ.loc[i,"거래소코드"]) == 2:
        print(i)
```

```python
# 연결 개별 데이터 프레임 합치기 (회사명, 거래소코드, 회계년도, 상장일, 상장폐지일, 소속코드를 기준으로)
KOSDAQ = KOSDAQ_1분기_연결.merge(KOSDAQ_1분기_개별, on = ['회사명', '거래소코드', '회계년도'], how = 'left')
```
