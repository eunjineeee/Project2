## 타겟 데이터 생성
- 부실인 경우 1, 부실 아닌 경우 0
- 상폐시점(t)이라면 상폐시점(t-1) 데이터를 1, 아닌 경우 0

```python
# 타겟 데이터인 부실여부 생성

KOSPI['부실여부'] = np.where((KOSPI['소속코드'] == 4) & (KOSPI['회계년도'].dt.year == KOSPI['상장폐지일'].dt.year -1), 1, 0)
KOSPI
```

## 부실 기업인데 과거 데이터가 타겟 값이 0인 경우 데이터 삭제

```python
# 부실 기업인데 과거 데이터가 타겟 값이 0인 경우 데이터 삭제

부실회사명 = KOSPI[KOSPI["부실여부"] == 1]['회사명'].unique()

for i in 부실회사명:
    drop_ind = KOSPI[(KOSPI['회사명'].isin(부실회사명)) & (KOSPI['부실여부'] == 0)].index
    KOSPI.drop(drop_ind, axis = 0, inplace = True)

KOSPI
```
